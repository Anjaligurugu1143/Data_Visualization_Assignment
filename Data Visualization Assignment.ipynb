# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
np.random.seed(42)
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = x1 * 0.5 + np.random.rand(n_samples) * 0.5
x3 = x1 * 0.2 + np.random.rand(n_samples) * 0.8
x4 = np.random.rand(n_samples)
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={"shrink": .8})
```

```python
import numpy as np
import pandas as pd
import plotly.express as px
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')
fig = px.imshow(pivot_table, 
                 title='Sales Heatmap by Month and Day', 
                 labels=dict(x="Month", y="Day", color="Average Sales"),
                 x=pivot_table.columns,
                 y=pivot_table.index)
fig.show()
```

# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')
show(p)
```

# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(5, 50, size=n)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', fill_alpha=0.6)
show(p)
```

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

# counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=800, height=400)
p.vbar(x=fruits, top=counts, width=0.5, color=factor_cmap('x', palette=['red', 'orange', 'yellow', 'green'], factors=fruits))
for fruit, count in zip(fruits, counts):
    p.text(fruit, count + 1, text=str(count), text_align='center')
show(p)
```

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           width=800, 
           height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black', alpha=0.7)
show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.

# data_heatmap = np.random.rand(10, 10)

# x = np.linspace(0, 1, 10)

# y = np.linspace(0, 1, 10)

# xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper
output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()
mapper = LinearColorMapper(palette="Viridis256", low=data_flat.min(), high=data_flat.max())
p = figure(title="Heatmap of Random Data", 
           x_axis_label='X', 
           y_axis_label='Y', 
           width=600, 
           height=600)
p.rect(x=x_flat, y=y_flat, width=0.1, height=0.1, 
        source={'x': x_flat, 'y': y_flat, 'color': data_flat}, 
        line_color=None, 
        fill_color=linear_cmap('color', mapper,low=1,high=5))
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
